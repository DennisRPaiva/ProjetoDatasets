A Segunda Guerra Mundial (1939-1945) representou um marco sombrio na história da humanidade, não apenas pelas atrocidades e devastação em larga escala, mas também pelo advento de uma nova era: a era nuclear. A explosão das bombas atômicas em Hiroshima e Nagasaki em 1945 pelos Estados Unidos revelou ao mundo o poder destrutivo inimaginável da energia nuclear.

Desde então, a atividade nuclear se disseminou por diversos países, impulsionada por diferentes motivações:

- Segurança nacional e dissuasão: O medo da proliferação nuclear e a busca por deterrentes contra ataques inimigos.
- Geração de energia: A promessa de uma fonte de energia limpa e abundante para suprir as demandas crescentes do mundo moderno.
- Aplicações médicas e industriais: O uso da tecnologia nuclear em diversas áreas como medicina, agricultura e pesquisa científica.

Este trabalho tem como objetivo analisar e comparar a atividade nuclear em diferentes países desde o fim da Segunda Guerra Mundial. Através de um estudo comparativo, busca-se:

- Identificar os principais países que desenvolveram programas nucleares e suas motivações.
- Analisar as diferentes trajetórias desses programas, desde a pesquisa e desenvolvimento até a proliferação nuclear e o desarmamento.
- Comparar os impactos da atividade nuclear em diferentes países, tanto em termos de segurança nacional quanto de desenvolvimento científico e tecnológico.
- Examinar os desafios e dilemas que a era nuclear apresenta para o mundo contemporâneo, como a proliferação nuclear.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Entendendo o dataset

Para ter uma ideia geral do dataset, será plotado abaixo um gráfico em linha do comportamento dos países em relação aos estoques nucleares a cada ano.

```python
directory = "nuclear_weapon"
file_path = os.path.join(directory, "nuclear_weapons_stockpiles.csv")

df = pd.read_csv(file_path)

grouped_data = df.groupby('country_name')

plt.figure(figsize=(10, 6))

for country, data in grouped_data:
    plt.plot(data['year'], data['nuclear_weapons_stockpile'], label=country)

plt.title('Estoque de Armas Nucleares por País ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Estoque de Armas Nucleares')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

O gráfico anterior mostra o estoque de armas nucleares por país ao longo do tempo. No entanto, EUA e Rússia possuem estoques significativamente maiores que os demais países. Isso distorce a visualização dos dados e dificulta a comparação entre os outros países.

Ao remover EUA e Rússia do gráfico, podemos:

* **Obter uma melhor visualização dos dados dos demais países:** O gráfico fica mais equilibrado e é possível observar com mais clareza as diferenças entre os estoques de armas nucleares dos demais países.
* **Comparar os demais países de forma mais justa:** Sem a presença de EUA e Rússia, é possível comparar os estoques de armas nucleares dos demais países de forma mais justa e precisa.
* **Evitar distorções na visualização dos dados:** A presença de EUA e Rússia no gráfico pode criar a falsa impressão de que esses países são os únicos com armas nucleares, o que não é verdade.

A década de 1980 representou um período crucial para o programa nuclear da África do Sul, marcado por desenvolvimentos clandestinos e crescente pressão internacional.

**Contexto Histórico:**

* Regime do Apartheid: Envolvido em conflitos internacionais e sob embargo de armas da ONU.
* Busca por autonomia estratégica e dissuasão nuclear.

**Desenvolvimentos Nucleares:**

* **Enriquecimento de Urânio:**
    * Instalação de plantas em Pelindaba e Valindaba.
    * Produção de urânio altamente enriquecido para armas nucleares.
* **Instalações de Teste:**
    * Kalahari: Testes de explosivos convencionais e desenvolvimento de mísseis balísticos.
    * Oceano Atlântico: Testes de dispositivos nucleares.
* **Arsenal Nuclear:**
    * Estimativa de seis armas nucleares montadas.
    * Desmantelamento gradual a partir de 1991.

**Pressão Internacional:**

* Isolamento diplomático e sanções econômicas.
* Preocupação com proliferação nuclear e regime do Apartheid.
* Negociações para desmantelamento do programa nuclear.

**Eventos Relevantes:**

* 1977: Embargo de armas da ONU.
* 1979: Explosão nuclear no Oceano Índico (atribuída à África do Sul).
* 1985: Ataque aéreo israelense à instalação de Osirak no Iraque (suspeita de colaboração com a África do Sul).
* 1991: F.W. de Klerk anuncia o desmantelamento do programa nuclear.
* 1994: Primeiras eleições democráticas na África do Sul.

**Consequências:**

* Desmantelamento completo do arsenal nuclear da África do Sul.
* Adesão ao Tratado de Não Proliferação Nuclear (TNP).
* Reconhecimento internacional como país livre de armas nucleares.

**Legado:**

* A atividade nuclear da África do Sul na década de 1980 levanta questões sobre ética, segurança e proliferação nuclear.

## Luta contra proliferação nuclear

Falando sobre a luta contra a proliferação nuclear, será que isso tem a ver com a queda brusca das principais nações nucleares (EUA, Rússia, Reino Unido e França) entre 1985 e 1995? Vamos descobrir. Abaixo, vamos usar o numpy para descobrir as taxas de queda dessas nações.

```python
# Selecionar os países e anos de interesse
df_first_world = df[(df['country_name'].isin(['United States', 'Russia', 'United Kingdom', 'France'])) &
        (df['year'].between(1985, 1995))]

# Criar coluna para armazenar derivadas
df_first_world['derivative'] = np.gradient(df_first_world['nuclear_weapons_stockpile'], df_first_world['year'])

# Agrupar os dados por país
grouped_data = df_first_world.groupby('country_name')
grouped_data.head()
```

```python
# Configurar o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Definir cores para cada país
colors = ['#0072B2', '#009688', '#4B0082', '#880E4F']

# Criar um iterador para as cores
color_iterator = iter(colors)

# Plotar a derivada para cada país
for country, data in grouped_data:
    # Obter a próxima cor da lista
    color = next(color_iterator)

    # Plotar a linha para o país
    plt.plot(data['year'], data['derivative'], color=color, label=country)

# Adicionar título e rótulos aos eixos
plt.title('Derivada do Estoque de Armas Nucleares (1985-1995)')
plt.xlabel('Ano')
plt.ylabel('Derivada do Estoque (Armas Nucleares/Ano)')

# Adicionar legenda
plt.legend(loc='best')

# Mostrar o grid
plt.grid(True)

# Ajustar o layout do gráfico
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

A partir destas derivadas nota-se que os Estados unidos e a Rússia têm derivadas muito negativas ao longo de boa parte do período destacado.

**Por que isso ocorreu?**

**Caso Russia:** A Rússia começa a reduzir a taxa de crescimento a partir de 1985, e começa a ter uma taxa negativa (ou seja, diminuir seus estoques) a partir do final deste mesmo ano. Isso é marcado pelo declínio da URSS, que passa a ter fortes crises econômicas, políticas e militares.

**Caso Estados Unidos:** Com o declínio soviético e o fim da Guerra Fria, a corrida armamentista perde o sentido. Com isso, os Estados Unidos param de investir em armas nucleares.

Outra forte motivação para o desarmamento nuclear foi a opinião pública e as tensões geopolíticas que poderiam surgir: uma guerra nuclear destruiria o planeta.

Mesmo com o fato de o investimento em armas nucleares ser diminuido drasticamente depois da Guerra Fria, Estados Unidos e Russia ainda são, disparadamente, as maiores potências nucleares do planeta.

Entretanto, podemos notar, a partir da análise destes gráficos, um desenvolvimento de alguns países de terceiro mundo em desu programas militares. O principal deles é a China, que no século atual ultrapassou Reino Unido e França e se tornou a terceira maior potência nuclear. 

## Investimento da China e de outros países de terceiro mundo

Abaixo será mostrado o crescimento de estoques nucleares de países de terceiro mundo.

```python
df_filtered = df[~df['country_name'].isin(['Russia', 'France', 'United States', 'United Kingdom'])]

grouped_data = df_filtered.groupby('country_name')

plt.figure(figsize=(10, 6))

for country, data in grouped_data:
    plt.plot(data['year'], data['nuclear_weapons_stockpile'], label=country)

plt.title('Estoque de Armas Nucleares por País ao Longo do Tempo (Excluindo China e Estados Unidos)')
plt.xlabel('Ano')
plt.ylabel('Estoque de Armas Nucleares')
plt.legend()  

plt.grid(True)
plt.tight_layout()
plt.show()
```

A partir deste gráfico, podemos ver mais claramente o comportamento da China. Também notamos o crecimento de duas nações importantes: Paquistão e Índia, andando lado a lado. Isso pode ser excplicado pelos motivos abaixo.

### Programa Nuclear da China:

- Início na década de 60, impulsionado por segurança nacional, desenvolvimento tecnológico e ideologia.
- Crescimento na década de 60 e 70, com testes nucleares e desenvolvimento de armas nucleares.
- Estagnação na década de 80, devido à distensão da Guerra Fria, priorização do desenvolvimento econômico e tratados de desarmamento.
- Ressurgimento a partir de 2010, com modernização do arsenal, investimento em novas tecnologias e expansão do arsenal.

### Crescimento dos Programas Nucleares da Índia e do Paquistão:

- Trajetória paralela e competitiva, iniciada nas décadas de 60 e 70.
- Motivada por disputas territoriais, rivalidade regional, Guerra Fria e proliferação nuclear.
- Desenvolvimento de arsenais nucleares com foco em mísseis balísticos de curto e médio alcance.
- Preocupações com risco de guerra nuclear, instabilidade regional e terrorismo nuclear.
- Esforços internacionais para pressionar os países a aderirem ao TNP e buscar a desnuclearização da região.
- Conclusão:

Os programas nucleares da China, Índia e Paquistão representam desafios à segurança regional e global.
Por mais que a comunidade internacional se mobilize para incentivar o diálogo, a diplomacia e a desnuclearização, os países que começam a desenvolver programas nucleares se sentem no direito de ter essas armas, assim como as maiores potências.
