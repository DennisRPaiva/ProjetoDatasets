```python
country_to_region = dict(zip(conflicts['Country'], conflicts['Region']))
```

```python
nuclear_stockpiles['Region'] = nuclear_stockpiles['country_name'].map(country_to_region)
nuclear_stockpiles.tail()
```

```python
nuclear_stockpiles
```

```python
nuclear_stockpiles[nuclear_stockpiles["nuclear_weapons_stockpile"] != 0]
```

```python
nuclear_nonzero = nuclear_stockpiles[nuclear_stockpiles["nuclear_weapons_stockpile"] != 0]
```

```python
periodo_pais_tem_arma_nuclear = nuclear_nonzero.groupby("country_name")["year"].agg(['first', 'last'])
periodo_pais_tem_arma_nuclear.head()
```

```python
#Primeiro: pega o primeiro pais da lista de 'periodo_pais_tem_arma_nuclear'
#Segundo: Retira todos os anos com 'BC' e 'AD'
#Terceiro: Retira todos os confitos com year1-year2 e transforma apenas para year1
#Quarto: pega todos as datas com um numero maior ou igual do que o primeiro ano de 'periodo_pais_tem_arma_nuclear' e menor e igual ultimo ano de 'periodo_pais_tem_arma_nuclear'


chinaConflicts = conflicts[conflicts["Country"].str.contains('China')]
chinaConflicts = chinaConflicts[~chinaConflicts['Date'].str.endswith(('BC', 'AD'))]
chinaConflicts['Date'] = chinaConflicts['Date'].str.slice(0, 4)
chinaConflicts['Date'] = chinaConflicts['Date'].astype(int)
chinaConflicts = chinaConflicts[(chinaConflicts["Date"] >= periodo_pais_tem_arma_nuclear.values[0][0]) & (chinaConflicts["Date"] <= periodo_pais_tem_arma_nuclear.values[0][1])]
chinaConflicts
```
